```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Example DataFrame (ensure your data is loaded correctly as a Pandas DataFrame)
# Replace this with your actual data
data = pd.DataFrame({
    'Geography': ['France', 'Spain', 'Germany', 'France', 'Spain']
})

# Initialize OneHotEncoder with sparse_output=False for dense output
one_hot_encoder_geo = OneHotEncoder(sparse_output=False)

# Ensure the column 'Geography' exists in the DataFrame and pass it as a 2D array
if 'Geography' in data.columns:
    geo_encoder = one_hot_encoder_geo.fit_transform(data[['Geography']])
else:
    raise ValueError("Column 'Geography' not found in the DataFrame.")

# Convert the encoded data into a Pandas DataFrame
geo_encoded_df = pd.DataFrame(
    geo_encoder,
    columns=one_hot_encoder_geo.get_feature_names_out(['Geography'])
)

# Concatenate the original DataFrame with the new one-hot encoded columns
data = pd.concat([data, geo_encoded_df], axis=1)

# Optionally, drop the original 'Geography' column
# data.drop('Geography', axis=1, inplace=True)

# Print the result
print(data)
```
